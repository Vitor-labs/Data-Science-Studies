```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, QuantileTransformer
```

```python
# Generate data with a long-tailed distribution (e.g., exponential distribution)
data = np.random.exponential(scale=1, size=1000).reshape(-1, 1)
# Create subplots for visual comparison
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original Data
axes[0].hist(data, bins=20, density=True)
axes[0].set_title("Original Data")

# Standard Scaler
scaler_std = StandardScaler()
data_std = scaler_std.fit_transform(data)
axes[1].hist(data_std, bins=20, density=True)
axes[1].set_title("Standard Scaler")

# Quantile Transformer
scaler_quantile = QuantileTransformer(output_distribution='normal')
data_quantile = scaler_quantile.fit_transform(data)
axes[2].hist(data_quantile, bins=20, density=True)
axes[2].set_title("Quantile Transformer (Normal Distribution)")

plt.tight_layout()
plt.show()
```
